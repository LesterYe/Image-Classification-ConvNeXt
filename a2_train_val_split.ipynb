```python
###### Libraries ######

import os
import shutil
import random
```

```python
###### Helpers ######

def train_val_split(dataset_dir, train_dir, val_dir):
  print(f'{101*"="}\nThe data splitting will take some time if your dataset is huge. Grab a coffee and come back later! :)\n{101*"="}')

  class_names = os.listdir(dataset_dir)
  num_classes = len(class_names)
  idx = 0
  for class_name in class_names:
    print(f'Class {idx}/{num_classes-1} : {class_name}')
    idx+=1
    dataset_class_dir = os.path.join(dataset_dir, class_name)
    train_class_dir = os.path.join(train_dir, class_name)
    val_class_dir = os.path.join(val_dir, class_name)
    try:
      os.makedirs(train_class_dir)
      os.makedirs(val_class_dir)
    except FileExistsError:
      pass

    image_names = os.listdir(dataset_class_dir)
    num_images = len(image_names)
    print(f'Number of images : {num_images}')

    count=0
    random.seed(seed)
    random.shuffle(image_names)
    for image_name in image_names:
      dataset_image_dir = os.path.join(dataset_class_dir, image_name)
      if count <= num_images*0.9: # 90/10 train/val split
        image_new_dir = os.path.join(train_class_dir, image_name) # train_image
      else:
        image_new_dir = os.path.join(val_class_dir, image_name) # val_image
      shutil.copy2(dataset_image_dir, image_new_dir)
      count+=1

    print(f'Splitting of {class_name} has completed.')
    num_train_images = len(os.listdir(train_class_dir))
    print(f'Number of train images : {num_train_images}')
    num_val_images = len(os.listdir(val_class_dir))
    print(f'Number of val images : {num_val_images}\n')
    
    # break
  print(f'{25*"="}\nTrain-Val Split Completed\n{25*"="}')
```

```python
###### Directories ######

dir = '/content/drive/MyDrive/CS424 - IP/a2/Image-Classification-ConvNeXt' # Change your working directory here
dataset_dir = os.path.join(dir, 'training', 'training') # Change your dataset directory here
train_dir = os.path.join(dir, 'train')
val_dir = os.path.join(dir, 'val')
```

```python
###### Change random seed for dataset random split ######
seed = 420 
```

```python
###### Dataset Train-Val Split ######
train_val_split(dataset_dir, train_dir, val_dir)
```
